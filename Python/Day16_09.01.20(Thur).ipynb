```python
#문제 2번 DashInsert
"""DashInsert 함수는 숫자로 구성된 문자열을 입력받은 뒤, 
문자열 내에서 홀수가 연속되면 두 수 사이에 - 를 추가하고, 
짝수가 연속되면 * 를 추가하는 기능을 갖고 있다. 
(예, 454 => 454, 4546793 => 454*67-9-3)
DashInsert 함수를 완성하자."""

New_list=[]
Num_list=list(map(int,input("Insert Number: ")))
for n in range(len(Num_list)-1):
    New_list.append(Num_list[n]) 
    if Num_list[n]%2==0 and Num_list[n+1]%2 ==0:
        New_list.append("*")
        
    if Num_list[n]%2==1 and Num_list[n+1]%2 ==1:
        New_list.append("-") 

New_list.append(Num_list[-1])
DashInsert = ''.join(New_list)
print(DashInsert)
    
    
```

```python
import numpy as np

n = input("number : ")


def DashInsert(n) :

    n_li = list(n)
    for i, x in enumerate(n_li):
        if x != '-' and x != '*':
          # 바로바로 *와 -을 추가할 것이므로 이 두 문자가 아닌 경우 -> 숫자(str)인 경우 가정
            x = int(x)
            try:
                if x % 2 == 0 and int(n_li[i+1])%2 == 0: # 짝수
                    n_li.insert(i+1, '*')
                elif x%2 != 0 and int(n_li[i+1])%2 != 0: # 홀수
                    n_li.insert(i+1, '-')
            except :
                pass
        

    string = ''.join(n_li)
    print(string)
    
c = DashInsert(n)
```
