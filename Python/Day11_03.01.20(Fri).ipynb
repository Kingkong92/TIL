```python
import requests
url="https://stackoverflow.com/questions/tagged/python"
response=requests.get(url)
#response.text # 엄청 더럽게 나온다. BeautifualSoup 깨끗하게 정리된다.
soup=BeautifulSoup(response.text, 'html.parser')
soup.select("div.question-summary h3 a") #h3그냥하면 밖에것도 다 나오고, > h3하면 h3안에 내용이 나온다.
links=[]

for link in soup.select("div.question-summary h3 a"):
    links.append(link.attrs['href'])
    #print(link.attrs['href']) #뒷부분 링크만
    #print("https://stackoverflow.com"+link.attrs['href']) #링크페이지로 연결

for link in links:
    url="https://stackoverflow.com"+link
    text=response.text
    soup=BeautifulSoup(text, 'html.parser') #Beau -> 역슬래쉬, 컴마, 이런거 정리
    
    print("Question\n")
    for x in soup.select("div.answercell div.post-text p"):
        print(x.text)
        
    print("\nAnswer\n")
    for x in soup.select("div.answercell div.post-text p"):
        print(x.text)
    print("="*40, "\n")
    
#     print("Question\n")
#     print("\nAnswer\n")
    
#     print("="*50,"\n")
```
