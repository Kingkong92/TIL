```python
import numpy as np
height=[0,0,3,5,2,4,9,0,6,4,0,6,0,0]
num=[-2,-1,1,2] # 2이상의 공간이 확보 되어있는 판단하기 위한 리스트
zero=np.zeros((15,15)) # 15x15 영행렬 생성
cnt=0
for i in height:
    if i !=0:
        zero[:14-i:-1,cnt]=1 # 영행렬에서 리스트 만큼의 높이의 빌딩 층수를 1로 찍어 줌 // 빌딩의 층 = 1
    cnt+=1
print(zero,"\n\n")    

dump=0
cnt=0

for i in range(15):
    for j in range(15):
        if zero[i][j]==1: # 행렬에서 1을 만나면
            cnt=0
            for k in num:
                if zero[i][j-k]==0:  # 1의 좌우 2칸이 전부 0인지 판단
                    cnt+=1
                if cnt==4:
                    dump+=1
# print(cnt)
print(dump)    
```

```python
# 예시 {0,0,3,5,2,4,8,0,6,4,0,6,0,0} 상황일때 조망권이 확보된 세대는 6
res=[]
num=[]
sample = [0,0,3,5,2,4,9,0,6,4,0,6,0,0]
for i in range(2,len(sample)-2):
    if sample[i]>sample[i-2] and sample[i]>sample[i-1] and sample[i]>sample[i+1] and sample[i]>sample[i+2]:
        res.append(i)
        num.append(sample[i]-max([sample[i-2],sample[i-1],sample[i+1],sample[i+2]]))

for i in range(len(res)):
    print("{}번째 빌딩에 {}세대".format(res[i],num[i]))
print("조망권이 보장된 세대는 총 %d세대입니다"%sum(num))
```
