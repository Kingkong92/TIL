```python
# 스칼라와 벡터/행렬간 곱셈

# 선형조합: 벡터/행렬에 스칼라값을 곱한후 더하거나 곱하는것
# 벡터/행렬의 선형 조합이라고 함.

#내적: 두 벡터의 곱셈(dot)
#두 벡터의 차원이 같아야 한다
#앞에 있는것이 행 벡터 * 뒤에 있는게 열 벡터
# 내적 결과는 스칼라
# Transpose 해줘야 한다.shape 맞춰줘야 한다.
```

```python
#내적의 example. dot 함수
x=np.array([[1],[2],[3]])
x      
y=np.array([[4],[5],[6]])            
y

#dot 함수. 내적
np.dot(x.T,y)
#array([[32]]) 2차원 배열

#@ dot. 내적
x.T@y
```

```python
#데이터벡터 * 가중치벡터 => 데이터벡터의 가중치 합.
#가중 평균. ex)학점별 가중치(1학점-100, 3학점-60)

"""
벡터의 내적은 두 벡터의 유사도를 계산하는데 
유사도? 두 벡터가 닮은 정도

코사인유사도?
"""
```

```python
#벡터간 유사도 계산. 
#d1과 각각의 벡터를 내적-> 스칼라 값.
digits=load_digits()

# 8*8 정사각형 그림
d1=digits.images[0]
d2=digits.images[10]
d3=digits.images[1]
d4=digits.images[11]

#64*1의 벡터 모양의 그림.
v1=d1.reshape(64,1)
v2=d2.reshape(64,1)
v3=d3.reshape(64,1)
v4=d4.reshape(64,1)

```

```python
plt.figure(figsize=(9,9)) # 1당 72 pixcell
import matplotlib.gridspec as gridspec
```

```python
v1 #'0'을 나타내는 이미지
v1.shape #(64, 1)
v1.T@v2 # v1과 v2 내적. 2차원 배열
(v1.T@v2)[0][0] #3064.0 # 스칼라 값을 얻을려면 2번 참조를 해야한다.
(v1.T@v3)[0][0] #1866.0
(v1.T@v4)[0][0] #1883.0
# 값이 높다고 해서(유사도가 높다고 해서) 똑같은 이미지가 아니다. 확률 상으로 높다는 뜻.
# 그림의 유사성은 CNN 기법을 사용. 
```
