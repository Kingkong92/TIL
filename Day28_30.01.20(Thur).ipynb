```python
titanic=sns.load_dataset('titanic')
titanic.info()
```

```python
import numpy as np
def count_missing(vec):
    nv=pd.isnull(vec)
    nc=np.sum(nv) #True:1, False:0
    return nc
    #누락값의 갯수 리턴
    #titanic.isnull(vec).sum()
    
def prop_missing(vec):
    #num=vec.size
    miss=count_missing(vec)
    return miss/vec.size

def prop_complete(vec):
    return 1-prop_missing(vec)
```

```python
#각 컬럼별 누락값의 개수가 모두 출력.
cm=titanic.apply(count_missing)
print(cm) 

#각 컬럼별 누락값의 비율을 모두 출력.
pm=titanic.apply(prop_missing)
print(pm)

#각 컬럼별 1에서 누락값의 뺀 수치.
pcm=titanic.apply(prop_complete)
print(pcm)
```

```python
#[1880df, 1881df, ...,2011df]
pieces=[]
for year in range(1880, 2011):
    path="C:/Users/student/Downloads/names/names/yob%d.txt." % year
    #print(path)
    df=pd.read_csv(path, names=['name','sex','births'])
    df['year']=year
    pieces.append(df)

    
names=pd.concat(pieces, ignore_index=True)
names    
```

```python
# names에서 year, sex
# ex) Mary는 1880년에 태어난 여자아이들 중 전체에서 차지하는 비중이 얼마인지 조사?

def add_prop(group):
    #births=group.births.sum()
    group['prop']=group.births/group.births.sum()
    return group
    #print(group.births)
names=names.groupby(['year','sex']).apply(add_prop)
names
#names.head(20)

# """
#     name sex births year prop
# 0   Mary  F   10000 1880 0.08
# """
```

```python
# 누적합 Cumsum -> 이름이 5명 이상의 데이터. => 이름의 다양성을 확인가능. 
prop_cumsum=df.sort_index(by='prop', ascending=False).prop.cumsum() # 84%만 동명이인. 나머지 16%는 5명 이상이 가지지 않는 이름. 
prop_cumsum[:10] #상위 10명에 이름에 대해. -> 9%나 차지.=> 해석: 현대에 들어서 이름의 다양성을 확인 가능. 
```

```python
#누적합이 0.5인 지점까지 이름 종류의 개수
prop_cumsum.values.searchsorted(0.5) #0 ~ 1 사이값 설정 # 50프로를 막 초과 하는 시점 기준.
#미국인 가지고 있는 이름의 50%가 116개 분류 안에 있다.
#이런식으로 분석주제 잡아서 분석할 수 있다. 
```

```python
# 'year','sex' 기준으로 그룹화
import numpy as np
def get_qc(group, q=0.5):
    group=group.sort_index(by='prop', ascending= False)
    return group.prop.cumsum().values.searchsorted(q)
    

#각 연도별, 상위 1000명, 그룹 year(연도별), sex(성별). 그룹화.
res=top1000.groupby(['year','sex']).apply(get_qc)
res
res=res.unstack('sex')
#res를 출력
res.plot()
```
