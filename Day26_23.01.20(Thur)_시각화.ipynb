```python
import numpy as np
import matplotlib.pyplot as plt


N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width, yerr=menStd)
p2 = plt.bar(ind, womenMeans, width,
             bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

```python
# 선(라인 플롯)
plt.title('plot')
plt.plot([10,20,30,40],[1,4,9,16],'rD:')
# plt.plot([ ]) []는 y이 값으로 하고 x는 1부터
#plt.plot([x축],[ y축])
# linestyle : 선 스타일 
# color : 색
# marker : 마크
#  rs-- 점선(-), rs-. : -.로 연결, rs: : 점선(.)
# rs : 네모 rh:원 rD: 마름모
plt.plot([10,20,30,40],[1,4,9,16],c = 'b', lw = 5, ls = '--', marker = 'o', ms = 15)
# c: color, lw: 선두께 ls:선스타일, marker: 마크스타일 ms: 마크 사이즈

plt.xlabel('x axis') # 축 제목
plt.ylabel('y axis')
plt.show()
```

```python
scPlot = plt.figure()
axes1 = scPlot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
```

```python
boxPlot = plt.figure()
axes1 = boxPlot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex'] =='Female']['tip'],
               tips[tips['sex'] == 'Male']['tip']],
              labels = ['Female','Male'])
```

```python
# 다변량 그래프
for i in tips[tips['sex']== 'Female'].index:
    tips.loc[i,'sex_color'] = 1
for i in tips[tips['sex']== 'Male'].index:
    tips.loc[i,'sex_color'] = 0

###### 다른 방법 1 #####################
tips['sex'].map({'Female':0,'Male':1})

###### 다른 방법 2 #####################
def enc(s):
    if s == 'Female':
        return 0
    else:
        return 1
tips['sex'].apply(enc)

###### 다른 방법 3 #####################
tips['sex'].apply(lambda x: 0 if x =='Female' else 1)
```

```python
sPlot = plt.figure()
axes1 = sPlot.add_subplot(1,1,1)
axes1.scatter(x=tips['total_bill'],
             y = tips['tip'],
             s = tips['size']*10, # 사이즈
             c = tips['sex_color'],# 색
             alpha = 0.5) # 투명도 1에 가까울수록 원색
```
